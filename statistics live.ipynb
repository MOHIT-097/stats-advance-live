## **Question7: Explain the Central Limit Theorem (CLT) with example.**

The Central Limit Theorem (CLT) is a fundamental statistical principle that describes the distribution of sample means. It states that, given a sufficiently large sample size, the distribution of the sample means will tend to be normally distributed, regardless of the original distribution of the population from which the samples are drawn. Here are the key aspects of the CLT:

### Key Points of the Central Limit Theorem

1. **Sample Size**: The theorem applies when the sample size is sufficiently large. A common rule of thumb is that a sample size of \( n \geq 30 \) is generally considered adequate.

2. **Independence**: The samples should be independent of one another.

3. **Convergence to Normality**: As the sample size increases, the distribution of the sample means approaches a normal distribution with mean \( \mu \) (the mean of the population) and variance \( \sigma^2/n \) (where \( \sigma^2 \) is the population variance).

### Example of the Central Limit Theorem

**Scenario**: Imagine you have a population of students at a university, and their exam scores are not normally distributed‚Äîlet's say they are skewed to the right (more students score lower with fewer high scores).

1. **Population Characteristics**:
   - Mean score (\( \mu \)): 70
   - Standard deviation (\( \sigma \)): 15
   - Distribution: Skewed right

2. **Sampling**: You take random samples of size \( n = 30 \) from this population and calculate the mean score for each sample.

3. **Repeated Sampling**: If you repeat this process many times (say, 1,000 samples), you will compile a distribution of the sample means.

4. **Observation**: According to the CLT:
   - Even though the original exam scores are not normally distributed, the distribution of the sample means will be approximately normally distributed.
   - The mean of this distribution of sample means will be close to the population mean (70).
   - The standard deviation of the sample means (standard error) will be \( \sigma/\sqrt{n} = 15/\sqrt{30} \approx 2.74 \).

5. **Conclusion**: After plotting the distribution of the sample means, you will see that it resembles a normal distribution, allowing you to use normal distribution techniques (e.g., confidence intervals, hypothesis testing) for inference, even if the original population was not normally distributed.

### Importance of the Central Limit Theorem

The CLT is crucial because it justifies the use of normal distribution methods in many statistical analyses, particularly when dealing with large samples. It allows researchers and statisticians to make inferences about population parameters based on sample statistics, facilitating hypothesis testing and confidence interval estimation.

## **Question8: Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be applied instead?**

Z-scores and t-scores are both standard scores used in statistics to describe how far a specific data point is from the mean in terms of standard deviations. However, they have different applications and are used under different conditions. Here‚Äôs a comparison:

### Z-Scores

- **Definition**: A z-score measures the number of standard deviations a data point is from the mean of a population.
- **Formula**:
  \[
  z = \frac{(X - \mu)}{\sigma}
  \]
  where \( X \) is the value, \( \mu \) is the population mean, and \( \sigma \) is the population standard deviation.
- **Use Case**:
  - **Population Known**: Use z-scores when the population standard deviation is known.
  - **Sample Size**: Typically applied when the sample size is large (\( n \geq 30 \)) due to the Central Limit Theorem, which suggests that the sampling distribution of the mean is approximately normal regardless of the original population distribution.
  - **Normal Distribution**: Ideal for data that follows a normal distribution.

### T-Scores

- **Definition**: A t-score measures the number of standard deviations a data point is from the sample mean, using the sample standard deviation.
- **Formula**:
  \[
  t = \frac{(X - \bar{X})}{s/\sqrt{n}}
  \]
  where \( \bar{X} \) is the sample mean, \( s \) is the sample standard deviation, and \( n \) is the sample size.
- **Use Case**:
  - **Population Unknown**: Use t-scores when the population standard deviation is unknown and must be estimated from the sample.
  - **Sample Size**: Typically applied for smaller sample sizes (\( n < 30 \)), where the sample might not adequately represent the population.
  - **Normality Assumption**: Used when the underlying population is assumed to be normally distributed, especially when the sample size is small.

### Summary of When to Use Each

- **Use Z-Scores**:
  - When the population standard deviation is known.
  - When the sample size is large (\( n \geq 30 \)).
  - When data is approximately normally distributed.

- **Use T-Scores**:
  - When the population standard deviation is unknown.
  - When the sample size is small (\( n < 30 \)).
  - When the underlying population is normally distributed, or the sample size is small enough that the t-distribution is more appropriate.

### Conclusion

In summary, z-scores and t-scores are both useful for standardizing scores, but they are applied under different circumstances. Understanding when to use each is crucial for accurate statistical analysis and inference.

## **Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?**

### **Task: Write Python code to calculate the z-score and p-value for the given data.**

### **Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.**

To calculate the z-score and p-value for the given data, we can use the following steps:

1. **Calculate the z-score** using the formula:
   \[
   z = \frac{(X - \mu)}{\sigma / \sqrt{n}}
   \]
   where:
   - \( X \) = sample mean (105)
   - \( \mu \) = population mean (100)
   - \( \sigma \) = population standard deviation (15)
   - \( n \) = sample size (25)

2. **Calculate the p-value** based on the z-score.

3. **Determine whether to reject or fail to reject the null hypothesis** based on the significance level of 0.05.

Here's how you can do this in Python:-

```python
import scipy.stats as stats
import numpy as np

# Given data
X = 105  # sample mean
mu = 100  # population mean
sigma = 15  # population standard deviation
n = 25  # sample size

# Calculate the z-score
z = (X - mu) / (sigma / np.sqrt(n))
print(f"Z-score: {z}")

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"P-value: {p_value}")

# Determine if we reject the null hypothesis
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

### Explanation of the Code:
- We first import the necessary libraries: `scipy.stats` for statistical functions and `numpy` for numerical operations.
- We define the variables based on the provided data.
- We calculate the z-score using the formula.
- We calculate the p-value using the cumulative distribution function (CDF) of the normal distribution.
- Finally, we check if the p-value is less than the significance level (0.05) to decide whether to reject or fail to reject the null hypothesis.

### Interpretation:
- The z-score will indicate how many standard deviations the sample mean is from the population mean.
- The p-value will tell us the probability of observing a sample mean as extreme as the one we have (or more extreme) under the null hypothesis. If the p-value is less than 0.05, we reject the null hypothesis; otherwise, we fail to reject it.

### Running the Code:
You can run this code in a Python environment with `numpy`, `matplotlib`, and `seaborn` installed to visualize the binomial distribution and verify the mean and variance properties.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.6  # probability of success
num_samples = 1000  # number of samples

# Generate binomial distribution samples
samples = np.random.binomial(n, p, num_samples)

# Calculate mean and variance
mean = np.mean(samples)
variance = np.var(samples)

# Expected mean and variance
expected_mean = n * p
expected_variance = n * p * (1 - p)

# Print results
print(f"Simulated Mean: {mean}")
print(f"Simulated Variance: {variance}")
print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance}")

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.histplot(samples, bins=np.arange(-0.5, n + 1.5, 1), kde=False, stat="density")
plt.title("Histogram of Binomial Distribution Samples")
plt.xlabel("Number of Successes")
plt.ylabel("Density")
plt.xticks(np.arange(0, n + 1, 1))
plt.grid(axis='y')
plt.show()
```
